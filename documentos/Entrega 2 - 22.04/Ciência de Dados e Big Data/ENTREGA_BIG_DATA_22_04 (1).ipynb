```python
# Garante que a coluna Schedule está no formato datetime
dfRide["Schedule"] = pd.to_datetime(dfRide["Schedule"])

# Cria colunas úteis para análise e predição
dfRide["Dia"] = dfRide["Schedule"].dt.weekday              # Dia da semana (0=segunda, 6=domingo)
dfRide["Hora"] = dfRide["Schedule"].dt.hour                # Hora inteira
dfRide["Minuto"] = dfRide["Schedule"].dt.minute            # Minutos
dfRide["HoraDecimal"] = dfRide["Hora"] + dfRide["Minuto"] / 60  # Hora como número decimal
dfRide["Faixa15min"] = dfRide["Schedule"].dt.floor("15min")     # Agrupamento de 15 em 15 minutos (útil pra clusters)

# Visualiza as novas colunas
dfRide.head(10)
```

```python
# Garante que os nomes das colunas estão padronizados
dfRideAdd = dfRideAdd.rename(columns={"RideAddressTypeID": "OrigDest"})

# Separa origem (OrigDest == 1)
dfOrigem = dfRideAdd[dfRideAdd["OrigDest"] == 1][["RideID", "Lat", "Lng"]].copy()
dfOrigem = dfOrigem.rename(columns={"Lat": "Lat1", "Lng": "Lng1"})

# Separa destino (OrigDest == 2)
dfDestino = dfRideAdd[dfRideAdd["OrigDest"] == 2][["RideID", "Lat", "Lng"]].copy()
dfDestino = dfDestino.rename(columns={"Lat": "Lat2", "Lng": "Lng2"})

# Junta origem e destino por RideID
dfCoordenadas = pd.merge(dfOrigem, dfDestino, on="RideID", how="inner")

# Arredonda coordenadas
dfCoordenadas["Lat1"] = dfCoordenadas["Lat1"].str.replace(",", ".").astype(float).round(6)
dfCoordenadas["Lng1"] = dfCoordenadas["Lng1"].str.replace(",", ".").astype(float).round(6)
dfCoordenadas["Lat2"] = dfCoordenadas["Lat2"].str.replace(",", ".").astype(float).round(6)
dfCoordenadas["Lng2"] = dfCoordenadas["Lng2"].str.replace(",", ".").astype(float).round(6)

# Adiciona apenas o Address de origem e destino
dfAddressOrig = dfRideAdd[dfRideAdd["OrigDest"] == 1][["RideID", "Address"]].rename(columns={"Address": "AddressOrig"})
dfAddressDest = dfRideAdd[dfRideAdd["OrigDest"] == 2][["RideID", "Address"]].rename(columns={"Address": "AddressDest"})

dfCoordenadas = pd.merge(dfCoordenadas, dfAddressOrig, on="RideID", how="left")
dfCoordenadas = pd.merge(dfCoordenadas, dfAddressDest, on="RideID", how="left")

# Visualizar resultado
dfCoordenadas.head()
```

```python
# RideID como string nas bases derivadas
dfTempo["RideID"] = dfTempo["RideID"].astype(str)
dfCoords["RideID"] = dfCoords["RideID"].astype(str)
dfEstimadaSelecionada["RideID"] = dfEstimadaSelecionada["RideID"].astype(str)

# Refiltra os dados com RideID comum
ids_comuns = set(dfTempo["RideID"]) & set(dfCoords["RideID"]) & set(dfEstimadaSelecionada["RideID"])

dfTempo = dfTempo[dfTempo["RideID"].isin(ids_comuns)].sort_values("RideID").reset_index(drop=True)
dfCoords = dfCoords[dfCoords["RideID"].isin(ids_comuns)].sort_values("RideID").reset_index(drop=True)
dfEstimadaSelecionada = dfEstimadaSelecionada[dfEstimadaSelecionada["RideID"].isin(ids_comuns)].sort_values("RideID").reset_index(drop=True)

# Concatena sem merge
dfDerivado = pd.concat([
    dfTempo,
    dfCoords.drop(columns=["RideID"]),
    dfEstimadaSelecionada.drop(columns=["RideID"])
], axis=1)

# Corrige possíveis vírgulas nas coordenadas e converte para float
for col in ["Lat1", "Lng1", "Lat2", "Lng2"]:
    dfDerivado[col] = dfDerivado[col].astype(str).str.replace(",", ".").astype(float).round(6)

# Calcula a distância geodésica entre origem e destino
from geopy.distance import geodesic
import numpy as np

Dist = []
for i in range(len(dfDerivado)):
    lat1 = dfDerivado.iloc[i]["Lat1"]
    lng1 = dfDerivado.iloc[i]["Lng1"]
    lat2 = dfDerivado.iloc[i]["Lat2"]
    lng2 = dfDerivado.iloc[i]["Lng2"]

    # Verifica se algum valor é nulo (por precaução)
    if np.isnan(lat1) or np.isnan(lng1) or np.isnan(lat2) or np.isnan(lng2):
        Dist.append(np.nan)
        continue

    ponto1 = (lat1, lng1)
    ponto2 = (lat2, lng2)
    dist_km = geodesic(ponto1, ponto2).kilometers
    Dist.append(dist_km)

# Adiciona a coluna de distância
dfDerivado["Distancia_km"] = Dist

# Visualiza com a nova coluna
dfDerivado.head()

```
